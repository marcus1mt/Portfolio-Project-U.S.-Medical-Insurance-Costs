# Find out the average age of the persons in the dataset.

```python
# Find out the average age of the persons in the dataset.
    
def find_average_age():
    average_age = 0
    ages = 0
    for num in insurance_cost['age']:
        ages += num
    average_age = ages / len(insurance_cost['age'])
    return average_age
```

# Plot Ages with Average Line


```python
x_axis = insurance_cost['charges']
y_axis = insurance_cost['age']
fig = plt.figure(figsize = (10, 5))
plt.bar(x_axis, y_axis, width=10)
plt.axhline(y= find_average_age(), color='r', linestyle = '-', label='Average Age')
plt.xlabel("Charges")
plt.ylabel("Age")
plt.title("Ages of Persons in Medical Insurance Data, With Average Age Line")
plt.ylim(15, 65)
plt.show()

```

```python
print('The average age of persons in this data set is: ' + str(find_average_age()))
```

# Look at the different costs between smokers vs. non-smokers.


```python
# Look at the different costs between smokers vs. non-smokers.
    
def find_smokers_cost_difference():
    smokers_cost = 0
    non_smoker_cost = 0
    s_count = 0
    ns_count = 0
    i = 0
    for smokes in insurance_cost['smoker']:
        if smokes == 'yes':
            smokers_cost += insurance_cost['charges'][i]
            s_count += 1
            i += 1
        elif smokes == 'no':
            non_smoker_cost += insurance_cost['charges'][i]
            ns_count +=1
            i += 1
    smokers_avgcost = (smokers_cost / s_count) 
    non_smokers_avgcost = (non_smoker_cost / ns_count)
    average_difference = smokers_avgcost - non_smokers_avgcost
    return smokers_avgcost, non_smokers_avgcost, average_difference
```

# Plot Smokers vs Non Smokers Average Charges and Difference"


```python
data = {'Avg Smokers Charge':find_smokers_cost_difference()[0], 
        'Avg Non Smokers Charge':find_smokers_cost_difference()[1], 
        'Ave Increase Charge for Smokers':find_smokers_cost_difference()[2]
       }
keys = list(data.keys())
values = list(data.values())
fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(keys, values, color ='maroon',
        width = 0.4)
 
plt.xlabel(" ")
plt.ylabel("Charges")
plt.title("Smokers vs Non Smokers Average Charges and Difference")
plt.show()
```

```python
print('The average charge for smokers is: ' + str(find_smokers_cost_difference()[0]))
print('The average charge for non smokers is: ' + str(find_smokers_cost_difference()[1]))
print('The average increased cost of a smoker is: ' + str(find_smokers_cost_difference()[2]))
```

# Figure out what the average cost for someone who has at least one child vs those with no children


```python
# Figure out what the average cost for someone who has at least one child vs those with no children
    
def find_average_with_children():
    cost_wkids = 0
    cost_count1 = 0
    cost_wokids = 0
    cost_count2 = 0
    i = 0
    for kids in insurance_cost['children']:
        if kids >= 1:
            cost_wkids += insurance_cost['charges'][i]
            cost_count1 +=1
            i += 1
        else:
            cost_wokids += insurance_cost['charges'][i]
            cost_count2 += 1
            i += 1           
    average_cost_with_children = cost_wkids / cost_count1
    average_cost_without_children = cost_wokids / cost_count2
    average_difference_increse_charges = average_cost_with_children - average_cost_without_children
    return average_cost_with_children, average_cost_without_children, average_difference_increse_charges
```

```python

print('The average charges of persons with children is: ' + str(find_average_with_children()[0]))
print('While the average charges of persons without children is: ' + str(find_average_with_children()[1]))
print('Thats only a small difference of: ' + str(find_average_with_children()[2]))
```

# Analyze where a majority of the individuals are from.

```python
# Analyze where a majority of the individuals are from.

def find_where_majority_is_from():
    nw = 0
    ne = 0
    sw = 0
    se = 0
    i = 0
    for home in insurance_cost['region']:
        if home == 'northwest':
            nw += 1
            i += 1
        elif home == 'northeast':
            ne += 1
            i += 1
        elif home == 'southwest':
            sw += 1
            i += 1
        else:
            se += 1
            i += 1
    return nw, ne, sw, se  
```

# Plot where the persons on data set are from

```python
data = {'People from Northwest':find_where_majority_is_from()[0], 
        'People from Northeast':find_where_majority_is_from()[1], 
        'People from Southwest':find_where_majority_is_from()[2],
        'People from Southeast':find_where_majority_is_from()[3]
       }
keys = list(data.keys())
values = list(data.values())
fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(keys, values, color ='red',
        width = 0.4)
plt.ylim(200, 400)
plt.xlabel(" ")
plt.ylabel("Number of People")
plt.title("Number of People Living in Each Region")
plt.show()
```

```python
print('People living in Nortwest Region: ' + str(find_where_majority_is_from()[0]))
print('People living in Norteast Region: ' + str(find_where_majority_is_from()[1]))
print('People living in Southwest Region: ' + str(find_where_majority_is_from()[2]))
print('People living in Southeast Region: ' + str(find_where_majority_is_from()[3]))
```
